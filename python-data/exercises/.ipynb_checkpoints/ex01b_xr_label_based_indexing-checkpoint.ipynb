```python
import xarray as xr
ds = xr.load_dataset('~/my-isc-work/python-data/data/xbhubo.pgc0apr.nc')
```

Q2. Select a subset of the `temperature` array using label-based indexing to get data at the position [0,0,0].

```python
temp_array = ds['temp']
temp_array = temp_array.squeeze(drop=True) #this drops any dimension with 0 or 1 value (in this case 't' for time)

temp_array.dims # call this to see what order the dimensions are in for indexing...
```

```python


temp_array[0,0,0] # this is position indexing

temp_array.isel(latitude=0, longitude=0, depth=0) # this is label-based indexing
```

Q3. Use `.loc` to find the temperature 5 meters below the sea surface in the South Atlantic where latitiude is -50.625 and longitude is 0.

```python
temp_array.loc[dict(latitude=-50.625, longitude=0, depth=5)] 
# this lets you look for data VALUES based on the variable names

#temp_array.loc[5,-50.625,0]
# this lets you look for data VALUES based on the variables, but not by name (so need to know the array order)

# need to call dict for 'dictionary' because we're not dealing with 2D data (?)
# using dimension names as dictionary keys so we can use loc with multidimension dataframe
```

Q4. It's not ideal to have to keep track of which dimension is in which position. Instead, use `.isel` to use the dimension names to get the data in the same place: this is depth position 0, latitude position 31 and longitude position 0.

```python
temp_array.isel(latitude=31, longitude=0, depth=0) #isel is by index 

temp_array.sel(latitude=-50.625, longitude=0, depth=5) #sel is by value in dimension
```

Q5. The previous method is still referring to a selection by integer position. Use `.sel` to give a labelled index with the named dimension to find the data at `lat=-78.5`, `lon=11.0`. You will need to pick a `method` too to select data with an inexact parameter.

```python
temp_array.sel(latitude=-50.5, longitude=0.0, method='nearest')
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
temp_array.sel(latitude=-50.5, longitude=0.0, method='nearest').plot(ax=ax)
ax.invert_yaxis()
ax.set_title("Temperature Profile at Lat=-50.5, Lon=0.0")
plt.show()
```
