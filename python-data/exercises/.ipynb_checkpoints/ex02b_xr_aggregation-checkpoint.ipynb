Q1. Import the `'../data/xbhubo.pgc0apr.nc'` dataset and create the temperature data array as in the last lesson.

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt

df = xr.open_dataset('~/my-isc-work/python-data/data/xbhubo.pgc0apr.nc')
df = df.squeeze()
```

```python
df_temp = df['temp']
df_temp
```

Q2. Let's compare the data between the sea surface and further down. Create two temperature datasets and extract the temperature change data the sea surface and the sea bottom

```python
df_temp.depth.min()
temp_shallow=df_temp.sel(depth=0, method='nearest')
# len(temp_shallow)
```

```python
df_temp.depth.max()
temp_deep = df_temp.sel(depth=5000, method = 'nearest')
# len(temp_deep)
```

Q3. Calculate the difference in temperature the bottom of the ocean and the surface.

```python
temp_diff = temp_deep - temp_shallow
temp_diff
```

Q4. Plot the difference in these temperatures using xarrays built-in features.

```python
fig, ax = plt.subplots()
temp_diff.plot(ax=ax)

# come back here to figure out how to change colourbar label 
```
